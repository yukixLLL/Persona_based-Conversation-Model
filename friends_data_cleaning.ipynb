```python
season_episode = {"01":24,"02":24,"03":25,"04":23,"05":23,"06":24,"07":23,
                  "08":23,"09":23,"10":17}
```

```python
for key in season_episode.keys():
    numepis = season_episode[key]
    for i in range(1,numepis+1):
        if i<10:
            filename ="clean_" + key + '0' + str(i)
        else:
            # 0213 doesn't exist because 13th episode has been included in 0212
            if key=='02' and i==13:
                continue
            if key=='06' and i==16:
                continue
            filename = "clean_" + key + str(i)
        with open(dir_path + filename + '.tsv') as f:
            lines = f.readlines()
            data_size = len(lines)
            print(data_size)
            for k in range(1,data_size):
                if ':' in lines[k] and 'dialogue' in lines[k]:
                    one_dia = lines[k].split('dialogue')[0]
                    speakers.append(one_dia.split(':')[0])
                    dialogue = one_dia.split(':')[1]
                    # remove action in the dialogue
                    dialogue = re.sub(r"\(.+\)",'',dialogue)
#                     print(dialogue)
                    dialogues.append(dialogue)
            print(filename + 'done')
```

```python
main_c = ['joey','rachel','chandler','monica','ross','phoebe']
```

```python
def separate(string, speaker):
    if ' ' in speaker:
        speaker = re.sub(' ','',speaker)
    speaker_list = speaker.split(string)
    for s in speaker_list:
        if s in main_c:
            speaker = s
            break
    return speaker

def main_speakers(speaker):
    speaker = speaker.lower()
    if "\"" in speaker:
        speaker = speaker.split("\"")[1]
    elif "&" in speaker:
        speaker = separate("&", speaker)
    elif "/" in speaker:
        speaker = separate("/", speaker)
    elif 'and' in speaker:
        if ',' in speaker:
            speaker = re.sub(',','and',speaker)
            speaker = separate("and", speaker)
        else:
            speaker = separate("and", speaker)
    elif ',' in speaker:
        speaker = separate(",", speaker)
    elif '[' and ']' in speaker:
        speaker = re.sub(r"\[.+\]",'',speaker)
    else:
        speaker_list = speaker.split(' ')
        if len(speaker_list) > 1:
            # adj. + name
            if speaker_list[-1] in main_c:
                speaker = speaker_list[-1]
            # with blank end
            elif speaker_list[0] in main_c:
                speaker = speaker_list[0]
    return speaker
```

```python
main_df['speakers'] = main_df['speakers'].apply(lambda x:main_speakers(x))
```
