```python
speakers = list()
dialogues = list()
```

```python
for key in season_episode.keys():
    numepis = season_episode[key]
    for i in range(1,numepis + 1):
        filename = key + '_'+ str(i)
        with open(dir_path + filename + '.txt') as f:
            lines = f.readlines()
            data_size = len(lines)
            print(data_size)
            for k in range(1,data_size):
                if ':' in lines[k] and 'Scene:' not in lines[k]:
#                     print(lines[k])
                    speaker = lines[k].split(':')[0]
                    dialogue = lines[k].split(':')[1]
                    # remove action in the dialogue
                    speaker = re.sub(r"\(.+\)",'',speaker)
                    speaker = re.sub(r"\[.+\]",'',speaker)
                    dialogue = re.sub(r"\(.+\)",'',dialogue)
                    dialogue = re.sub(r"\[.+\]",'',dialogue)
                    dialogue = dialogue.lstrip(' ')
#                     print(dialogue)
                    speakers.append(speaker)
                    dialogues.append(dialogue)
            print(filename + 'done')
```

```python
def speaker_processor(speaker):
    flag =0
    speaker = speaker.lower()
    speaker = speaker.strip(' ')
    if 'and' in speaker or '&' in speaker:
        if 'together' in speaker:
            speaker = speaker.rstrip(' together')
        if ',' in speaker:
            speaker = re.sub(',','and',speaker)
        if '&' in speaker:
            speaker = re.sub('&','and',speaker)
        if ' ' in speaker:
            speaker = re.sub(' ','',speaker)
        speaker_list = speaker.split('and')
        for s in speaker_list:
            if s in main_c:
                speaker = s
                break
    else:
        speaker_list = speaker.split(' ')
        if len(speaker_list) > 1:
            # adj. + name
            # number. blank name
            if speaker_list[-1] in main_c:
                speaker = speaker_list[-1]
            elif speaker_list[0] in main_c:
                speaker = speaker_list[0]
    if speaker == 'rajj':
        speaker = 'raj'
    if speaker =='\xa0sheldon' or 'sheldon-bot' in speaker:
        speaker ='sheldon'
    if speaker =='\xa0howard':
        speaker = 'howard'
    return speaker
```

```python
big_bang_df['speakers'] = big_bang_df['speakers'].apply(lambda x: speaker_processor(x))
```
