```python
import re
```

```python
speakers = list()
dialogues = list()
```

```python
with open('data/clean_friends/clean_1001.tsv') as f:
    lines = f.readlines()
    data_size = len(lines)
    print(data_size)
    for k in range(1,data_size):
        if 'dialogue' in lines[k]:
            one_dia = lines[k].split('dialogue')[0]
            speakers.append(one_dia.split(':')[0])
            dialogue = one_dia.split(':')[1]
            # remove action in the dialogue
            dialogue = re.sub(r"\(.+\)",'',dialogue)
            print(dialogue)
            dialogues.append(dialogue)
```
